```python
import numpy as np
import random as r
import matplotlib.pyplot as plt
```

```python
def emp(n):
    muestra=[]
    for i in range(n):
        u1=r.random()
        u2=r.random()
        if u1>u2:
           muestra.append(u1)
    return muestra
```

```python
L=emp(3000)
```

```python
plt.hist(L, bins=50)
```

```python
a=0
b=3
c=5
M=2/(c-a)
def triangular(x):
    if a<=x<=b:
       return 2*(x-a)/((c-a)*(b-a))
    elif b<x<=c:
       return 2*(c-x)/((c-a)*(c-b))
    else:
       return 0
```

```python
def empt(n):
    muestra=[]
    for i in range(n):
        u1=r.random()
        u2=r.random()
        x=a+(c-a)*u1
        if u2<=triangular(x)/M:
           muestra.append(x)
    return muestra
```

```python
T=empt(3000)
```

```python
plt.hist(T, bins=50)
```

```python
def fa(x):
    if 0<=x<=1:
       return 3/4
    elif 1<x<=5/3:
       return (15-9*x)/8
    else:
       return 0
```

```python
Ma=3/4
def empa(n):
    muestra=[]
    for i in range(n):
        u1=r.random()
        u2=r.random()
        x=(5/3)*u1
        if u2<=fa(x)/Ma:
           muestra.append(x)
    return muestra
```

```python
Ia=empa(30000)
```

```python
plt.hist(Ia, bins=50)
```

```python
def fb(x):
    if 4<=x<=5:
       return (11/4-x/2)
    elif 5<x<=6:
       return (x/2-9/4)
    else:
       return 0
```

```python
Mb=3/4
def empb(n):
    muestra=[]
    for i in range(n):
        u1=r.random()
        u2=r.random()
        x=(4+2*u1)
        if u2<=fb(x)/Mb:
           muestra.append(x)
    return muestra
```

```python
Ib=empb(30000)
```

```python
plt.hist(Ib, bins=50)
```

```python
def f2(x):
    if -1<=x<=1:
       return (3*x**2)/2
    else:
       return 0
```

```python
M2=3/2
def emp2(n):
    muestra=[]
    for i in range(n):
        u1=r.random()
        u2=r.random()
        x=(-1+2*u1)
        if u2<=f2(x)/M2:
           muestra.append(x)
    return muestra
```

```python
I2=emp2(30000)
```

```python
plt.hist(I2, bins=50)
```
