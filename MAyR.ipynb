```python
import numpy as np
import random as r
import matplotlib.pyplot as plt
```

```python
def emp(n):
    muestra=[]
    for i in range(n):
        u1=r.random()
        u2=r.random()
        if u1>u2:
           muestra.append(u1)
    return muestra
```

```python
L=emp(3000)
```

```python
plt.hist(L, bins=50)
```

```python
a=0
b=3
c=5
M=2/(c-a)
def triangular(x):
    if a<=x<=b:
       return 2*(x-a)/((c-a)*(b-a))
    elif b<x<=c:
       return 2*(c-x)/((c-a)*(c-b))
    else:
       return 0
```

```python
def empt(n):
    muestra=[]
    for i in range(n):
        u1=r.random()
        u2=r.random()
        x=a+(c-a)*u1
        if u2<=triangular(x)/M:
           muestra.append(x)
    return muestra
```

```python
T=empt(3000)
```

```python
plt.hist(T, bins=50)
```
