```python
import random as r
import numpy as np
import matplotlib.pyplot as plt
```

```python
def triangular(n):
    lis=[]
    for i in range(n):
        u=r.random()
        if 0<=u<0.5:
           x=np.sqrt(2*u)
           lis.append(x)
        else:
            x=2-np.sqrt(2*(1-u))
            lis.append(x)
    return lis
```

```python
R=triangular(1000)
```

```python
plt.hist(R)
```

```python
def cdf(M):
    n=len(M)
    x=np.sort(M)
    y=np.arange(0,n)/n
    return x,y
```

```python
x_math,y_math=cdf(R)
```

```python
media=np.mean(R)
```

```python
plt.step(x_math, y_math)
plt.axhline(y=0.5, color='red')
plt.axvline(x=media, color='green')
```

```python
def we(n):
    lis=[]
    for i in range(n):
        u=r.random()
        x=(-np.log(1-u))**(1/2)
        lis.append(x)
    return lis
```

```python
w=we(10000)
```

```python
plt.hist(w, bins=30, density=True)
plt.show()
```

```python
def gu(n):
    lis=[]
    for i in range(n):
        u=r.random()
        x=-np.log(-np.log(u))
        lis.append(x)
    return lis
```

```python
g=gu(10000)
```

```python
plt.hist(g, bins=30, density=True)
plt.show()
```
