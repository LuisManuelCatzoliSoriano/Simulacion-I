```python
import numpy as np
import scipy.stats as stats
import requests
import ast
import matplotlib.pyplot as plt
import tabulate
import random as r
```

```python
def numa(n):
    numeros=[]
    for i in range(n):
     a=r.random()
     numeros.append(a)
    return numeros
```

```python
numa(5)
```

```python
nu=numa(100)
bins=[0,0.2,0.45,0.7,0.9,1.01]
indices=np.digitize(nu, bins, right=False)
conteos=[(indices == i).sum() for i in range(1, len(bins))]
frecuencias, intervalos=np.histogram(nu, bins)
print("Frecuencias:", frecuencias)
print("Intervalos:", intervalos)
```
