```python
import random as r
import numpy as np
import matplotlib.pyplot as plt
```

```python
def beneficio(ventas, compras=400):
    if ventas<compras:
        beneficio=ventas*1000+(compras-ventas)*250-compras*750
    else:
        beneficio=compras*1000-compras*750
    return beneficio
```

```python
beneficio(300)
```

```python
def Venta():
    venta=r.random()
    if 0<=venta<0.22:
       U_ven=100
    elif 0.22<=venta<0.47:
       U_ven=200
    elif 0.47<=venta<0.58:
       U_ven=300
    elif 0.58<=venta<0.83:
       U_ven=400
    elif 0.83<=venta<0.94:
       U_ven=500
    elif 0.94<=venta<=1:
       U_ven=600
    return U_ven
```

```python
Venta()
```
