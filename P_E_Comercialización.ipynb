```python
import random as r
import numpy as np
import matplotlib.pyplot as plt
```

```python
def beneficio(ventas, compras=400):
    if ventas<compras:
        beneficio=ventas*1000+(compras-ventas)*250-compras*750
    else:
        beneficio=compras*1000-compras*750
    return beneficio
```

```python
beneficio(300)
```

```python
def venta():
    venta=r.random()
    if 0<=venta<0.22:
       U_ven=100
    elif 0.22<=venta<0.47:
       U_ven=200
    elif 0.47<=venta<0.58:
       U_ven=300
    elif 0.58<=venta<0.83:
       U_ven=400
    elif 0.83<=venta<0.94:
       U_ven=500
    elif 0.94<=venta<=1:
       U_ven=600
    return U_ven
```

```python
venta()
```

```python
def Ben_l(n):
  bene=[]
  prom=[]
  for i in range(n):
    v=venta()
    b=beneficio(v)
    bene.append(b)
    p=np.sum(bene)/(i+1)
    prom.append(p)
  return prom
```

Se observo en clase que para $n=850$, se volvió estable:

```python
l=Ben_l(850)
plt.plot(l)
```

```python
np.mean(l)
```

Con el experimento anterior fijamos el tamaño de la muestra del experimento de simulación.

```python
def simu(n):
  simu=[]
  for i in range(n):
    b=np.mean(Ben_l(850))
    simu.append(b)
  return simu
```

```python
def p_hi(n):
 s=simu(n)
 h=plt.hist(s, bins=10)
 p=np.mean(s)
 ds=np.std(s)
 ps=print("El promedio es:", p)
 ds=print("La desviación estándar es:", ds)
 return h,ps
```
