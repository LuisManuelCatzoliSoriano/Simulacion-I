```python
import numpy as np
import random as r
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.sqrt(np.arctan(x))
```

```python
f(1)
```

```python
xx=np.linspace(0,1,100)
plt.plot(xx, f(xx))
```

```python
def mcc(a, b, N):
    I=0
    for i in range(N):
        x=r.random()
        I+=f(x)
    return I/N
```

```python
mcc(0,1,1000)
```

```python
def experimento(M):
    L=[]
    for i in range(M):
        I=mcc(0,1,i+1)
        L.append(I)
    return L
```

```python
E=experimento(600)
```

```python
plt.plot(E)
```

```python
np.mean(E)
```

```python
def simulacion(F):
    Li=[]
    for i in range(F):
        e=np.mean(experimento(600))
        Li.append(e)
    return Li
```

```python
p=simulacion(100)
```

```python
plt.hist(p)
```
