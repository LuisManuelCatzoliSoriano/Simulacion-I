```python
import random as r
import math
import matplotlib.pyplot as plt

def laplace(n):
    lis = []

    for i in range(n):

        # Elegir componente
        u1 = r.random()

        # Uniforme para la exponencial
        u2 = r.random()

        # Exponencial(1)
        y = -math.log(u2)

        # Composición
        if u1 <= 0.5:
            x = y
        else:
            x = -y

        lis.append(x)

    return lis
```

```python
muestra = laplace(1000)
plt.hist(muestra, bins=50)
```

```python
def trapezoidal_derecha(a, n):

    lis = []

    for i in range(n):

        u1 = r.random()
        u2 = r.random()

        # Parte uniforme
        if u1 <= a:
            x = u2

        # Parte con densidad 2x
        else:
            x = math.sqrt(u2)

        lis.append(x)

    return lis
```

```python
muestra2 = trapezoidal_derecha(0.2,10000)
plt.hist(muestra2, bins=50)
```
