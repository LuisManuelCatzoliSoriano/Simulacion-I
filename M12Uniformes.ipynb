```python
import numpy as np
import random as r
import matplotlib.pyplot as plt
```

```python
def gaussian(mu, sigma, n):
    muestra=[]
    for i in range(n):
        suma=0
        for j in range(12):
            u=r.random()
            suma=suma+u
        x=mu+sigma*(suma-6)
        muestra.append(x)
    return muestra
```

```python
M=gaussian(0,1,100000)
```

```python
plt.hist(M, bins=50, density="True")
plt.grid()
plt.show()
```
