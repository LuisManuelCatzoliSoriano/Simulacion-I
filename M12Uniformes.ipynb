```python
import numpy as np
import random as r
import matplotlib.pyplot as plt
```

```python
def gaussian(mu, sigma, n):
    muestra=[]
    for i in range(n):
        suma=0
        for j in range(12):
            u=r.random()
            suma=suma+u
        x=mu+sigma*(suma-6)
        muestra.append(x)
    return muestra
```

```python
M=gaussian(0,1,100000)
```

```python
plt.hist(M, bins=50, density="True")
plt.grid()
plt.show()
```

```python
def muller(n):
    muestra2=[]
    for i in range(n):
        u1=r.random()
        u2=r.random()
        x=np.sqrt(-2*np.log(u1))*np.cos(2*np.pi*u2)
        y=np.sqrt(-2*np.log(u1))*np.sin(2*np.pi*u2)
        muestra2.append(x)
        muestra2.append(y)
    return muestra2
```

```python
K=muller(100000)
```

```python
plt.hist(K, bins=50, density="True")
```
